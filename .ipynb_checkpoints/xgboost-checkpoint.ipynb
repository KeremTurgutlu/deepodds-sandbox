```python
def sort_dict(d, reverse=True):
    import operator
    return sorted(d.items(), key=operator.itemgetter(1), reverse=reverse)

def col_nunique(df): 
    """get number of unique elements"""
    d = {c: df[c].nunique() for c in df.columns}
    return sort_dict(d)
    
def col_nas(df): 
    """get number of NAs"""
    d = {c:df[c].isnull().sum() for c in df.columns}
    return sort_dict(d)
```

```python
import xgboost as xgb
import random

class XGBModelData:
    def __init__(self, trn_df, val_df, trn_y, val_y, test_df=None):
        
        self.trn_df = trn_df
        self.val_df = val_df
        self.trn_y = trn_y
        self.val_y = val_y
        self.test_df = test_df
        
    
    def get_train_eval_ds(self):
        """Data for evaluation mode"""
        dtrain = xgb.DMatrix(self.trn_df, label=trn_y)
        dval = xgb.DMatrix(self.val_df, label=val_y)
        evals =  [(dtrain, 'train'), (dval, 'valid')]
        return dtrain, dval, evals
    
    def get_train_test_ds(self):
        """Data for final training mode"""
        full_trn_df = pd.concat([self.trn_df, self.val_df])
        full_trn_y = pd.concat([pd.Series(self.trn_y), pd.Series(self.val_y)])
        
        dtrain = xgb.DMatrix(full_trn_df, label=full_trn_y)
        dtest = xgb.DMatrix(self.test_df)
        return dtrain, dtest
    
        
    @classmethod
    def from_cv_idxs(cls, df, y, cv_idxs):
        """
        df (pd.DataFrame): Dataframe for independent variable
        y (pd.Series, np.array): Dependent variable
        cv_idxs (list): list of indexes/bool masks for trn, val, test sets
        """
        
        if len(cv_idxs) == 3: test_df = df[cv_idxs[-1]] 
        else: test_df = None
        trn_idxs, val_idxs = cv_idxs[:2]
        trn_df, val_df, trn_y, val_y = df[trn_idxs], df[val_idxs], y[trn_idxs], y[val_idxs]        
        return cls(trn_df, val_df, trn_y, val_y, test_df)
```

```python
class XGBLearner:
    def __init__(self, dtrain, evals, maximize, eta, objective, metric, seed=None):
        self.maximize = maximize
        self.eta = eta
        self.objective = objective
        self.metric = metric
        self.dtrain = dtrain
        self.evals = evals
        if seed is not None: self.seed = seed


        params = {}
        # model init
        params['objective'] = self.objective
        params['eta'] = self.eta
        params['eval_metric'] = self.metric
        params['seed'] = 20
        params['num_class'] = 3
        # model complexity
        params['max_depth'] = 5
        params['min_child_weight'] = 10
        params['gamma'] = 1
        # noise and variance
        params['subsample'] = 1
        params['colsample_bytree'] = 1
        params['colsample_bylevel'] = 1

        # run to find best iteration
        model = xgb.train(params=params,
                          dtrain=self.dtrain,
                          evals=self.evals,
                          maximize=self.maximize,
                          num_boost_round=10//self.eta,
                          early_stopping_rounds=1//(10*self.eta),
                          verbose_eval=False)
        self.num_rounds = model.best_iteration


    def max_depth_generator(limits):
        """upper and lower limits"""
        return NotImplementedError

    def random_params(self, seed=None):
        """
        Generates random parameters for random search
        """
        if seed is not None: random.setstate(seed)

        max_depth = random.randint(3,20) # max depth
        min_c_w = random.randint(1,100) # min child weight
        gamma = random.randint(1,10) # gamma
        subsample = random.uniform(0,1) # subsample
        colsample_bytree = random.uniform(0,1) # colsample by tree
        colsample_bylevel = random.uniform(0,1) # colsample by level

        params = {}
        params['objective'] = self.objective
        params['eval_metric'] = self.metric
        params['eta'] = self.eta 
        params['seed'] = self.seed
        params['num_class'] = 3 # fix this for non-classification cases
        # model complexity
        params['max_depth'] = max_depth
        params['min_child_weight'] = min_c_w
        params['gamma'] = gamma
        # noise and variance
        params['subsample'] = subsample
        params['colsample_bytree'] = colsample_bytree
        params['colsample_bylevel'] = colsample_bylevel
        return params

    def random_search(self, n_search, verbose=False):
        """Do random search for n_search times""" 
        self.best_params = None
        self.best_iteration = None

        for i in range(n_search):
            # get new random params
            params = self.random_params()

            model = xgb.train(params=params, 
                              dtrain=self.dtrain,
                              evals=self.evals,
                              maximize=self.maximize,
                              num_boost_round=int(self.num_rounds*2),
                              early_stopping_rounds=self.num_rounds//2,
                              verbose_eval=False)


            params_score, params_iteration = model.best_score, model.best_iteration

            if i == 0: best_score = params_score

            if (self.maximize & (params_score > best_score)) | (~self.maximize & (params_score < best_score)):
                best_score = params_score
                self.best_iteration = params_iteration
                self.best_params = params
                if verbose: print(f"Found new best score {best_score}")

        print("----------------------------------------")
        print("Best params and best iteration are found")
        print("----------------------------------------")

    def fit_best_model(self, dtrain):
        """fits best model based on random search"""
        self.best_model = xgb.train(params=self.best_params, 
                                    dtrain=dtrain,
                                    num_boost_round=self.best_iteration,
                                    verbose_eval=False)


    def predict(self, dtest): return self.best_model.predict(dtest)
```
